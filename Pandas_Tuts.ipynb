#### Group by and Aggregating

```python
df = pd.read_csv(r"D:/data/PandasYouTubeSeries-main/flavors.csv")
df
```

```python
grouped_by_base = df.groupby('Base Flavor')
```

```python
grouped_by_base.mean()
```

```python
df.groupby('Base Flavor').mean()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'count', 'sum'], 'Texture Rating' : ['mean', 'max', 'count', 'sum']})
```

```python
df.groupby(['Base Flavor', 'Liked']).mean()
```

#### Merge, Join, Concatenate

```python
df1 = pd.read_csv(r"D:/data/PandasYouTubeSeries-main/LOTR.csv")

df1
```

```python
df2 = pd.read_csv(r"D:/data/PandasYouTubeSeries-main/LOTR 2.csv")
df2
```

```python
#df1.merge(df2) 
  
df1.merge(df2, how ='inner', on='FellowshipID') 
```

```python
df1.merge(df2, how ='outer') 
```

```python
df1.merge(df2, how ='cross') 
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_left', rsuffix= '_right')

#Join better with indexes
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix='_right', how='inner')

df4
```

```python
#Concatenate

pd.concat([df1, df2], join ='inner')
```
